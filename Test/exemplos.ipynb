````python
### Describing a numeric Series
s = pd.Series([1, 2, 3])
s.describe()
count    3.0 # Número total de elementos na série categórica
mean     2.0 # a media da soma total dos valores 
std      1.0 # Medida de dispersão que indica o quão longe os valores do conjunto estão da média
min      1.0 # o menor valor 
25%      1.5 # O valor que divide os 25% menores valores da série numérica para esquerda
50%      2.0 # O valor que divide a série numérica ao meio
75%      2.5 # O valor que divide os 75% menores valores da série numérica para direita
max      3.0 # o maior valor
dtype: float64

count: Número total de elementos em um conjunto de dados. Indica a contagem total de observações ou itens.

mean: (média): A média aritmética dos valores em um conjunto de dados. Obtida somando todos os valores e dividindo pelo número total de elementos.

std: (desvio padrão): Medida de dispersão que indica o quão longe os valores do conjunto estão da média. Um desvio padrão mais alto indica maior dispersão.

min: (mínimo): O valor mais baixo em um conjunto de dados, representando o menor elemento.

25%: (primeiro quartil): O valor que divide os 25% menores valores da série numérica. Também é conhecido como primeiro quartil ou quartil inferior.

50%: (mediana): O valor que divide a série numérica ao meio, deixando 50% dos valores abaixo e 50% acima. Também é conhecido como mediana.

75%: (terceiro quartil): O valor que divide os 75% menores valores da série numérica. Também é conhecido como terceiro quartil ou quartil superior.

max: (máximo): O valor mais alto em um conjunto de dados, representando o maior elemento.
````

````python
# Remover os valores NaN e converter para int, tratando valores não válidos
# Codigo
data['Year'] = pd.to_numeric(data['Year'], errors='coerce', downcast='integer')

    - data['Year']: Acessa a coluna 'Year' no DataFrame 'data'.

    - pd.to_numeric(...): 
        - Esta função do pandas é utilizada para converter argumentos para números. 
        - Neste caso, ela está sendo usada para converter os valores na coluna 'Year' para numéricos.

    - data['Year'] = pd.to_numeric(...): 
        - Atribui os valores convertidos de volta à coluna 'Year' no DataFrame 'data'.

    - errors='coerce': 
        - Este argumento especifica que, se houver algum valor na coluna 'Year' que não possa ser convertido para um número, ele deve ser substituído por NaN (Not a Number). 
        - Isso é útil para lidar com valores não numéricos ou valores ausentes.

    - downcast='integer': 
        - Este argumento indica que, se possível, o pandas deve downcast (reduzir a precisão) dos valores numéricos para o menor tipo de dados possível. 
        - Neste caso, está sendo especificado 'integer', o que significa que a pandas tentará representar os números como inteiros, se possível.

````

````python
# Filtrar os dados para do período de 2015 a 2018
filtered_data = data[(data['Year'] >= 2015) & (data['Year'] <= 2018)]
    - filtered_data esta recebendo duas condicionais 
        - Se DF 'Year' for maior que 2015 e se DF for menor que 2018
````

````python
# Identificar o valor da coluna com a maior frequência para cada ano
    # Legenda das colunas
        - coluna_principal = 'Species '
        - coluna_referencia = 'Year'
        - Count = 'index_redefinido'

# Codigo        
top_species_by_year = (
    filtered_data.groupby(['coluna_referencia', 'coluna_principal'])
    .size()
    .reset_index(name='index_redefinido')
    .sort_values(['coluna_referencia', 'index_redefinido'], ascending=[True, False])
    .drop_duplicates('coluna_referencia')
)

    - filtered_data.groupby(['coluna_referencia', 'coluna_principal']).size(): 
        - O conjunto de dados filtered_data é agrupado por duas colunas, 'Year' e 'Species', usando o método groupby. 
        - A função size() é aplicada para contar o número de ocorrências em cada combinação única de ano e espécie. 
        - O resultado é uma Série que mostra a contagem de ocorrências para cada combinação de ano e espécie.

    - .reset_index(name='index_redefinido'): 
        - O índice do DataFrame resultante da operação groupby é redefinido, e a Série resultante é convertida de volta em um DataFrame. 
        - A nova coluna que contém a contagem é nomeada como 'Count'.

    - .sort_values(['coluna_referencia', 'index_redefinido'], ascending=[True, False]): 
        - O DataFrame é ordenado primeiro pela coluna 'Year' em ordem ascendente e, em seguida, pela coluna 'Count' em ordem descendente. 
        - Isso é feito para que, dentro de cada ano, a espécie com a contagem mais alta apareça primeiro.

    - .drop_duplicates('coluna_referencia'): 
        - A função drop_duplicates é usada para manter apenas a primeira ocorrência de cada ano, removendo as duplicatas. 
        - Isso significa que para cada ano, apenas a espécie com a contagem mais alta será mantida no DataFrame final.

````

````python
# Criar um gráfico de barras
barplot = sns.barplot(x="Year", y="Count", hue="Species ", data=top_species_by_year)

# Adicionar rótulos e título
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Species with Highest Frequency of attacks from 2015 to 2018')

# Rotacionar rótulos do eixo x para melhor visualização (opcional)
plt.xticks(rotation=45, ha='right')

# Exibir o gráfico
plt.show()


    - barplot = sns.barplot(x="Year", y="Count", hue="Species ", data=top_species_by_year): 
        - Um gráfico de barras é criado usando a função barplot do Seaborn. 
        - Os argumentos 'x', 'y', e 'hue', especificam as variáveis para o eixo x, eixo y e a diferenciação de cores (nesse caso, as diferentes espécies), respectivamente. 
        - Os dados utilizados são do DF top_species_by_year, que foi previamente criado para conter a espécie com a maior frequência para cada ano.

    - plt.xlabel('Species'): 
        - Adiciona um rótulo ao eixo x, indicando que as barras representam diferentes espécies.

plt.ylabel('Count'): 
    - Adiciona um rótulo ao eixo y, indicando que as alturas das barras representam a contagem de ataques.

plt.title('Species with Highest Frequency of attacks from 2015 to 2018'): 
    - Adiciona um título ao gráfico, fornecendo uma descrição geral do que está sendo representado.

plt.xticks(rotation=45, ha='right'): 
    - Rotaciona os rótulos do eixo x em 45 graus para melhorar a legibilidade. 
    - O argumento ha='right' alinha os rótulos à direita para evitar sobreposições.

plt.show(): Exibe o gráfico.
````
