# Como pegar os valores de uma tabela 

```python
# verificando a classe da data
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
```

```python
# Verificando todas as tabelas 
list_table = soup.find_all('table')
list_table
```

```python
# Verificando as tables por classe
list_table = soup.find_all('table')
list_table = soup.find('table', class_='wikitable')
list_table
```

```python
header_2021 = soup.find_all('span', attrs={'id':"Street_League_Skateboarding_Super_Crown,_Men's:_Jacksonville,_FL,_USA"})
header_2021[0]['id'].replace('_', ' ')
```

```python
# contruindo um dataframe
df = pd.DataFrame(columns=['Rank', 'Skater', 'Score'])
df
```

```python
# pegando os valores da table, limpando e passando pro dataframe
data_list = []
for row in list_table.tbody.find_all('tr'): 
    columns = row.find_all('td')
    if columns:
        rank = columns[0].text.strip()
        skater = columns[1].text.strip()
        score = columns[2].text.strip()
        score = float(BeautifulSoup(score, 'html.parser').text) if score else None

        # Adiciona um novo dicionário à lista
        data_list.append({'Rank': rank, 'Skater': skater, 'Score': score})

# Cria o DataFrame a partir da lista de dicionários
df = pd.DataFrame(data_list)

# Imprime o DataFrame
print(df)
```
